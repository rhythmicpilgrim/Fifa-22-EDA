```python
# Importing Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('players_22.csv',low_memory = False)
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='overall', y='potential')

# Add title and axis labels
plt.title('Overall Rating vs. Potential Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Potential Rating')

# Show the plot
plt.show()
```

```python
sns.displot(data=df, x='overall', kde=True)
```

```python
import seaborn as sns

sns.set(style="whitegrid")
sns.barplot(x="short_name", y="overall", data=df[:10], palette="Blues_d")
```

```python
#This bar plot shows the top 10 players by overall rating. From the graph, we can see that Lionel Messi has the highest overall rating among all the players, followed by Robert Lewandowski and Cristiano Ronaldo.
```

```python


sns.set(style="whitegrid")
sns.scatterplot(x="potential", y="value_eur", data=df, hue="age", palette="magma")
```

```python
#This scatter plot shows the relationship between potential and market value of players, with age as the hue. We can see that younger players with higher potential ratings tend to have higher market values, while older players with lower potential ratings have lower market values.
```

```python

sns.set(style="whitegrid")
sns.lineplot(x="age", y="overall", data=df, hue="player_positions", palette="Set2")
```

```python
#This line plot shows the overall rating of players at different ages, with player positions as the hue. We can see that players who play as strikers and center forwards tend to have higher overall ratings at a younger age, while defenders and goalkeepers tend to have higher overall ratings at an older age.
```

```python

corr = df.corr()
sns.heatmap(corr, cmap="YlGnBu")
```

```python
# This heatmap shows the correlation between different numerical variables in the dataset. We can see that there is a strong positive correlation between overall and potential ratings, while age has a weak positive correlation with potential rating.
```

```python
sns.boxplot(x="player_positions", y="value_eur", data=df, palette="Set3")
```

```python
#This box plot shows the distribution of market values across different player positions. We can see that forwards and attacking midfielders tend to have higher market values compared to defenders and goalkeepers.
```

```python
sns.violinplot(x="age", y="value_eur", data=df, palette="magma")
```

```python
#This violin plot shows the distribution of market values across different ages. We can see that players in their mid-20s tend to have higher market values compared to younger and older players.
```

```python
sns.histplot(df["age"], kde=False, color="purple")
```

```python
#This histogram shows the distribution of player ages in the dataset. We can see that the majority of players are in their mid-20s to early 30s.
```

```python

sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
ax = sns.scatterplot(x="value_eur", y="potential", data=df, hue="player_positions", palette="magma")
ax.set_xlabel("Value (in millions of Euros)", fontsize=12)
ax.set_ylabel("Potential Rating", fontsize=12)
ax.set_title("Player Value vs. Potential", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#This plot shows the relationship between player value and potential. There is a positive correlation between these two variables, meaning that players with higher potential tend to have higher value. Strikers and attacking midfielders tend to have the highest value and potential ratings
```
